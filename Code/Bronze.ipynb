## *Moving file from Volum Source_data to Bronze layer*

```python
import os
Volumm_Souce_files=os.listdir("/Volumes/bara_slaes_project/bronze/source_data")
```

```python
df=(
    spark.read.option("header",True)
    .option("inferSchema",True)
    .csv("/Volumes/bara_slaes_project/bronze/source_data/dim_customers.csv")
)

df.write.mode("overwrite").saveAsTable("bara_slaes_project.bronze.dim_customers")
```

```python
df=(
    spark.read.option("header",True)
    .option("inferSchema",True)
    .csv("/Volumes/bara_slaes_project/bronze/source_data/dim_products.csv")
)

df.write.mode("overwrite").saveAsTable("bara_slaes_project.bronze.dim_products")
```

```python
df=(
    spark.read.option("header",True)
    .option("inferSchema",True)
    .csv("/Volumes/bara_slaes_project/bronze/source_data/fact_sales.csv")
    
)

df.write.mode("overwrite").saveAsTable("bara_slaes_project.bronze.fact_sales")
```
