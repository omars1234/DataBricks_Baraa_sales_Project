## *Moving file from Volum Source_data to Bronze layer*

```python
INGESTION_CONFIG = [
    {
        "source": "crm",
        "path": "/Volumes/bara_slaes_project/bronze/source_data/source_crm/cust_info.csv",
        "table": "crm_cust_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/bara_slaes_project/bronze/source_data/source_crm/prd_info.csv",
        "table": "crm_prd_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/bara_slaes_project/bronze/source_data/source_crm/sales_details.csv",
        "table": "sales_details"
    }
    
    ]

for item in INGESTION_CONFIG:
    #print(f"Ingesting {item['source']} â†’ workspace.bronze.{item['table']}") 
    print(item["path"])
    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"bara_slaes_project.bronze.{item['table']}")
    )
```
