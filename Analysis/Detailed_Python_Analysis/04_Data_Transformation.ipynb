```python
pyspark_df=spark.read.table("bara_slaes_project.gold.selected_features_table")
```

```python
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml import Pipeline
from pyspark.sql.functions import rand, col, lit, monotonically_increasing_id
import pyspark.sql.functions as F

x=pyspark_df.drop("Sales")
stages = []

assembler_inputs =  x.columns
stages += [VectorAssembler(inputCols=assembler_inputs, outputCol="features")]

full_pipeline = Pipeline(stages=stages)
model = full_pipeline.fit(pyspark_df)
df_ready = model.transform(pyspark_df)
df_ready.select("features").show(truncate=False)
```
