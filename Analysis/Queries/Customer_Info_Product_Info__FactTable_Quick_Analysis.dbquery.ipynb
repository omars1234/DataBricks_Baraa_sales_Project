```sql
SELECT 
cust.marital_status,
cust.Gender,
cust.Country,
fact.Sales,
fact.Order_Date,
prod.Category_Name,
prod.Sub_Category_Name,
prod.Product_Line,
prod.Product_name
 FROM bara_slaes_project.gold.dimcustomer_infO AS cust
LEFT JOIN bara_slaes_project.gold.factable AS fact
ON cust.Customer_ID=fact.Customer_ID
LEFT JOIN bara_slaes_project.gold.dimproduct_info AS prod
ON fact.Product_ID=prod.Product_ID
```
